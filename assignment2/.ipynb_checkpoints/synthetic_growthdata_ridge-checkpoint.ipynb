```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, r2_score

# Generate the Growth Data Set
np.random.seed(42)
x = np.linspace(0,1,20).reshape(-1,1)
y = np.exp(1.5*x) + 0.5*np.random.rand(x.size,1)

plt.scatter(x,y,color='b')
plt.grid()
plt.show()
```

```python
# Polynomial degree
degree = 10

# Train Linear Regression Model
model_linear = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model_linear.fit(x, y)
y_linear_pred = model_linear.predict(x)

# Train Ridge Regression Model
alpha = 0.1
model_ridge = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=alpha))
model_ridge.fit(x, y)
y_ridge_pred = model_ridge.predict(x)
```

```python
# Plot the results
plt.scatter(x, y, color='b', label='Data')
plt.plot(x, y_linear_pred, color='r', label='Linear Regression')
plt.plot(x, y_ridge_pred, color='g', label=f'Linear Regression w/ Ridge (alpha={alpha})')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute performance metrics for Linear Regression
r2_linear = r2_score(y, y_linear_pred)
mse_linear = mean_squared_error(y, y_linear_pred)
nmse_linear = mse_linear / np.var(y)  # Normalized MSE

# Compute performance metrics for Ridge Regression
r2_ridge = r2_score(y, y_ridge_pred)
mse_ridge = mean_squared_error(y, y_ridge_pred)
nmse_ridge = mse_ridge / np.var(y)  # Normalized MSE

# Print results
print("Performance Metrics for Linear Regression Model:")
print(f"R^2: {r2_linear:.4f}")
print(f"MSE: {mse_linear:.4f}")
print(f"NMSE: {nmse_linear:.4f}\n")

print("Performance Metrics for Ridge Regression Model:")
print(f"R^2: {r2_ridge:.4f}")
print(f"MSE: {mse_ridge:.4f}")
print(f"NMSE: {nmse_ridge:.4f}")
```

```python
# Extract coefficients for Linear Regression
linear_coefficients = model_linear.named_steps['linearregression'].coef_
linear_intercept = model_linear.named_steps['linearregression'].intercept_

# Extract coefficients for Ridge Regression
ridge_coefficients = model_ridge.named_steps['ridge'].coef_
ridge_intercept = model_ridge.named_steps['ridge'].intercept_

# Print the coefficients
print("Linear Regression Model Coefficients:")
print(f"Intercept: {linear_intercept}")
print(f"Coefficients: {linear_coefficients}\n")

print("Ridge Regression Model Coefficients:")
print(f"Intercept: {ridge_intercept}")
print(f"Coefficients: {ridge_coefficients}")
```
